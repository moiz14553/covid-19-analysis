```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
covid_df
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head(4)
```

```python
#ACTIVE CASES
#in this we add new column which is active cases by adding cured cases with active cases then subtract from confirmed cases
covid_df['Active_Cases']=(covid_df['Confirmed'])-(covid_df['Cured'])+(covid_df['Deaths'])
covid_df.tail(5)
```

```python
statewise=pd.pivot_table(covid_df,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise['Recovery rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')#pivot table with styling
```

```python
#top 10 active cases states
top_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title('top 10 states with active cases in india',size=25)
ax=sns.barplot(data=top_active_cases.iloc[:10],y="Active_Cases",x="State/UnionTerritory",linewidth=2,edgecolor='red')

plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
#top states with highest death rate
top_10_death=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
plt.figure(figsize=(18,5))
plt.title("top 10 states with most deaths rate",size=25)
ax=sns.barplot(data=top_10_death.iloc[:12],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel('states')
plt.ylabel('total death cases')
plt.show()
```

```python
#Growth trend
fig=plt.figure(figsize=(12,6))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh']),x='Date',y='Active_Cases',hue='State/UnionTerritory'])
ax.set_title('top 5 affected states in India',size=16)
```

```python
fig = plt.figure(figsize=(12,6))

ax = sns.lineplot(
    data=covid_df[covid_df['State/UnionTerritory'].isin(
        ['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh']
    )], 
    x='Date', 
    y='Active_Cases', 
    hue='State/UnionTerritory'
)

ax.set_title('Top 5 Affected States in India', size=16)

# Format date labels
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))  # Show every 3rd month
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format as 'Jan 2021'

plt.xticks(rotation=45)  # Rotate x-axis labels
plt.show()
```
